```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
colors = ["#FF6F61", "#6B5B95"]
```

```python
data_path = 'data.csv'
df = pd.read_csv(data_path)
df.head()
```

```python
print("Cancelled bookings (P_c):", df['P-C'].sum())
print("Non-cancelled bookings (P_not_C):", df['P-not-C'].sum())
```

```python
df['booking status'] = df['booking status'].map({'Not_Canceled': 0, 'Canceled': 1})

market_mapping = {
    'Online': 0,
    'Offline': 1, 
    'Corporate': 2,
    'Aviation': 3,
    'Complementary': 4
}
df['market segment type'] = df['market segment type'].map(market_mapping)

meal_mapping = {
    'Meal Plan 1': 0,
    'Not Selected': 1,
    'Meal Plan 2': 2,
    'Meal Plan 3': 3
}
df['type of meal'] = df['type of meal'].map(meal_mapping)

room_mapping = {
    'Room_Type 1': 0,
    'Room_Type 2': 1,
    'Room_Type 3': 2,
    'Room_Type 4': 3,
    'Room_Type 5': 4,
    'Room_Type 6': 5,
    'Room_Type 7': 6
}
df['room type'] = df['room type'].map(room_mapping)
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap", fontsize=16)
plt.tight_layout()
plt.savefig("plots/correlation_heatmap.png", dpi=300)
plt.show()
```

```python
# Count plot of cancellations

plt.figure(figsize=(6,4))
sns.barplot(x=["Cancelled", "Not Cancelled"], y=[df['P-C'].sum(), df['P-not-C'].sum()], palette=colors)
plt.title("Total Booking Behavior", fontsize=16)
plt.ylabel("Number of Bookings")
plt.tight_layout()
plt.savefig('plots/total_booking_behavior.png')
plt.show()
```

```python
total_bookings = df.shape[0]

canceled = df[df['booking status'] == 1].shape[0]

cancel_percent = (canceled / total_bookings) * 100

print(f"Canceled bookings: {canceled} out of {total_bookings}  {cancel_percent:.2f}%")
print("===============")
labels = ['Canceled', 'Not Canceled']
sizes = [canceled, total_bookings - canceled]
colors = ['#FF4136', '#2ECC40']

plt.figure(figsize=(5,5))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title("Booking Cancellation Overview")
plt.tight_layout()
plt.savefig("plots/booking_cancellation_pie.png", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='repeated', palette="coolwarm")
plt.title("Are Guests Repeating Their Visits?", fontsize=16)
plt.xlabel("Repeated Guest")
plt.ylabel("Count")
plt.tight_layout()
plt.savefig('plots/repeated_guests.png')
plt.show()
```

```python
plt.figure(figsize=(8,8))
df['market segment type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title("Market Type Share", fontsize=16)
plt.ylabel("")  # Hide y-label
plt.tight_layout()
plt.savefig('plots/market_type_share.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['lead time'], bins=40, kde=True, color="#FF6F61")
plt.title("Distribution of Lead Time", fontsize=16)
plt.xlabel("Days before Arrival")
plt.ylabel("Number of Bookings")
plt.tight_layout()
plt.savefig('plots/lead_time_distribution.png')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=(df['P-C'] > 0), y='lead time', data=df, palette=colors)
plt.title("Do Long Lead Times Lead to Cancellations?", fontsize=16)
plt.xlabel("Was Cancelled (True = Cancelled)")
plt.ylabel("Lead Time (days)")
plt.tight_layout()
plt.savefig('plots/lead_time_vs_cancellation.png')
plt.show()
```

```python
cancel_rate_by_market = df.groupby('market segment type')['P-C'].mean().sort_values()

plt.figure(figsize=(10,6))
sns.barplot(x=cancel_rate_by_market.values, y=cancel_rate_by_market.index, palette="rocket")
plt.title("Average Cancellation Rate by Market Type", fontsize=16)
plt.xlabel("Cancellation Rate")
plt.ylabel("Market Type")
plt.tight_layout()
plt.savefig('plots/cancellation_rate_by_market.png')
plt.show()
```

```python
# Are certain market types booking waaay early, or are they last-minute types?

plt.figure(figsize=(10,6))
sns.boxplot(data=df, x='market segment type', y='lead time', palette="Spectral")
plt.title("Lead Time by Market Type", fontsize=16)
plt.xticks(rotation=30)
plt.ylabel("Lead Time (Days)")
plt.tight_layout()
plt.savefig("lead_time_by_market_type.png", dpi=300)
plt.show()
```

```python
# Do new guests ghost you more often than loyal ones?

cancel_by_repeat = df.groupby('repeated')['P-C'].mean()

plt.figure(figsize=(6,4))
sns.barplot(x=cancel_by_repeat.index.astype(str), y=cancel_by_repeat.values, palette=colors)
plt.title("Cancellation Rate: New vs Repeated Guests", fontsize=16)
plt.xlabel("Is Repeated Guest")
plt.ylabel("Average Cancellation Rate")
plt.tight_layout()
plt.savefig("cancellation_by_repeated.png", dpi=300)
plt.show()
```

```python
pivot = df.pivot_table(values='P-C', index='market segment type', columns='repeated', aggfunc='mean')

sns.set(font_scale=1.1)
sns.set_style("whitegrid")

plt.figure(figsize=(8, 6))
ax = sns.heatmap(pivot, annot=True, fmt=".2f", cmap="Reds", linewidths=.5, cbar_kws={'label': 'Avg P-C (Cancellations)'})
plt.title("Avg Cancellations by Market Type & Guest Loyalty", fontsize=16, weight='bold')
plt.xlabel("Repeated Guest (0 = First Time, 1 = Returning)", fontsize=12)
plt.ylabel("Market Segment Type", fontsize=12)
plt.tight_layout()
plt.savefig("heatmap_market_repeated_pc.png", dpi=300)
plt.show()
```

```python
selected_features = ['lead time', 'P-C', 'P-not-C', 'special requests', 'average price ']
corr_features = df[selected_features]

corr = corr_features.corr()

plt.figure(figsize=(8, 6))
sns.set(style="whitegrid")

ax = sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="vlag",
    linewidths=0.5,
    center=0,
    cbar_kws={'label': 'Correlation'}
)

plt.title("Feature Correlation Heatmap", fontsize=16, weight='bold')
plt.tight_layout()
plt.savefig("plots/correlation_heatmap_vlag.png", dpi=300)
plt.show()
```

```python
# Do long lead times really mean theyll cancel more often?

plt.figure(figsize=(8,5))
sns.scatterplot(x='lead time', y='P-C', data=df, alpha=0.4, color='#0081a7')
plt.title("Scatter: Lead Time vs Cancellation Count", fontsize=16)
plt.ylabel("Cancellation Count")
plt.tight_layout()
plt.savefig("plotsleadtime_vs_cancellation_scatter.png", dpi=300)
plt.show()
```

```python
volume_cancellation = df.groupby('market segment type').agg({'P-C':'mean', 'P-not-C':'count'}).sort_values(by='P-C', ascending=False)

fig, ax1 = plt.subplots(figsize=(10,6))
sns.barplot(x=volume_cancellation.index, y=volume_cancellation['P-C'], ax=ax1, color='#e76f51')
ax1.set_ylabel("Cancellation Rate", color='#e76f51')
ax1.set_title("Cancellation Rate and Booking Volume by Market Type")
ax1.tick_params(axis='x', rotation=30)

ax2 = ax1.twinx()
sns.lineplot(x=volume_cancellation.index, y=volume_cancellation['P-not-C'], ax=ax2, color='#264653', marker="o")
ax2.set_ylabel("Booking Volume", color='#264653')

plt.tight_layout()
plt.savefig("plots/cancel_rate_vs_volume.png", dpi=300)
plt.show()
```

```python
# Number of Adults and Children
print("=== GUEST NUMBERS ===")
print(f"Average adults per booking: {df['number of adults'].mean():.1f}")
print(f"Average children per booking: {df['number of children'].mean():.1f}")
print(f"Bookings with children: {(df['number of children'] > 0).sum()} ({(df['number of children'] > 0).mean()*100:.1f}%)")

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
df['number of adults'].hist(bins=10, color='#FF6F61', alpha=0.7)
plt.title('Number of Adults')
plt.xlabel('Adults')

plt.subplot(1, 3, 2)
df['number of children'].hist(bins=10, color='#6B5B95', alpha=0.7)
plt.title('Number of Children')
plt.xlabel('Children')

plt.subplot(1, 3, 3)
(df['number of adults'] + df['number of children']).hist(bins=15, color='#2ecc71', alpha=0.7)
plt.title('Total Family Size')
plt.xlabel('Total People')

plt.tight_layout()
plt.savefig('plots/guest_numbers.png', dpi=300)
plt.show()
```

```python
# Night Stays Analysis
print("\n=== STAY DURATION ===")
print(f"Average weekend nights: {df['number of weekend nights'].mean():.1f}")
print(f"Average week nights: {df['number of week nights'].mean():.1f}")
total_nights = df['number of weekend nights'] + df['number of week nights']
print(f"Average total nights: {total_nights.mean():.1f}")

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
df['number of weekend nights'].hist(bins=10, color='#e74c3c', alpha=0.7)
plt.title('Weekend Nights')
plt.xlabel('Nights')

plt.subplot(1, 3, 2)
df['number of week nights'].hist(bins=15, color='#3498db', alpha=0.7)
plt.title('Week Nights')
plt.xlabel('Nights')

plt.subplot(1, 3, 3)
total_nights.hist(bins=20, color='#9b59b6', alpha=0.7)
plt.title('Total Nights')
plt.xlabel('Nights')

plt.tight_layout()
plt.savefig('plots/night_stays.png', dpi=300)
plt.show()
```
